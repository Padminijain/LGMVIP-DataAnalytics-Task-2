### TASK- 2 : Image to Pencil Sketch with Python:


### Import the required libraries

```python
from PIL import Image
from IPython.display import display
import cv2
```

#### Load the respective image

```python
image = cv2.imread('ABoy.jpg', 1)
image = cv2.resize(image, (640, 600))
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
display(Image.fromarray(image))
```

#### Current image to greyscale conversion

```python
image_grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(image_grey))
```

#### Transformation of the greyscale image into negative using bitwise_not operator

```python
image_inv = cv2.bitwise_not(image_grey)
display(Image.fromarray(image_inv))
```

```python
image_smooth = cv2.GaussianBlur(image_inv, (21, 21), 0) # To smoothe the imahe using Gaussian function
display(Image.fromarray(image_smooth))
```

```python
def dodge(image_grey, image_smooth):
  image_smooth_inv = 255 - image_smooth
  return cv2.divide(image_grey, image_smooth_inv, scale=256.0)
```

#### Final pencil sketch of the respective image

```python
image_final = dodge(image_grey, image_smooth)
display(Image.fromarray(image_final))
```

#### Compare the original picture with its pencil sketch

```python
display(Image.fromarray(image))
display(Image.fromarray(image_final))
```
